# Load the Clean Dataset and New Dataset

- **Clean Survey Dataset**: The student survey contains comprehensive information about students studying at BCIT, including demographic details, financial data, lifestyle choices, and time allocation. 
- **2024 Salary By Age Group Dataset**: Contains average salary data based on age groups from Year 2024. (**Fake Data** - file generated by gen_fake_data.ipynb saved in **Scripts Folder**

```python
import pandas as pd

import os
print(os.getcwd())

# Load the clean survey dataset last week: Saved in Clean_Data Folder
df_clean = pd.read_csv("../Clean_Data/clean_survey_2025-02-06_assign4_sosochoi.csv")

# Load the fake new dataset: Saved in Data Folder
df_salary = pd.read_csv("../Data/2024_salarybyagegroup_data.csv")
```

# Display the Clean Dataset Before Merging

```python
print("Shape of Clean Data:", df_clean.shape)
df_clean
```

# Display the 2024 Salary By Age Group Dataset Before Merging

```python
print("Shape of Fake Data:", df_salary.shape)
df_salary
```

# Rename Columns for 2024 Salary By Age Group Dataset 

```python
# Rename columns
df_salary.rename(columns={
    "Average Salary ($)": "Average Salary ($) by Age Group",
    "Median Salary ($)": "Median Salary ($) by Age Group",
    "Salary Range ($)": "Salary Range ($) by Age Group",
    "Employment Rate (%)": "Employment Rate (%) by Age Group"
}, inplace=True)

# Display the DataFrame
print(df_salary)
```

# Merge the Datasets

```python
# Merge the datasets
merged_df = pd.merge(df_clean, df_salary, on="Age Group", how="left")
# Display the shape of the merged DataFrame
print("Shape of Merged DataFrame:", merged_df.shape)
merged_df
```

# Explain the Type of Merge Chosen: Left Join

- **Data Retention**: To retain all records from the clean dataset (which has 250 rows) and include corresponding salary information based on age groups.
- **Handling Missing Data**: To ensures that if some rows in the clean dataset do not fall into any of the defined age groups in the salary dataset, those records will still be included, with NaN filling in the salary-related fields.

# Type of NA Values are Created 

- The salary-related fields for individuals not falling into defined age groups will contain NaN values. 
- This provides clarity in the dataset, indicating that while the demographic data is available, the corresponding salary information is missing.
- From the merged dataset, we could see that the age group "Under 20" has missing salary data.

# Save the Merged Dataset

```python
# Save the merged dataset: saved in **"Results" Folder**

merged_df.to_csv("../Results/merged_student_data.csv", index=False)
```
